```python
#extract only values for a specific date
fifth_feb = data[data['date']=='02/05/2019']
```

```python
fifth_feb= fifth_feb.rename(columns={'Longitude':'lon','Latitude':'lat'})
```

```python
#initializing distance and duration function
def get_distance(point1: dict, point2: dict) -> tuple:
    """Gets distance between two points en route using http://project-osrm.org/docs/v5.10.0/api/#nearest-service"""
    
    url = f"""http://router.project-osrm.org/route/v1/driving/{point1["lon"]},{point1["lat"]};{point2["lon"]},{point2["lat"]}?overview=false&alternatives=false"""
    r = requests.get(url)
    
    # get the distance from the returned values
    route = json.loads(r.content)["routes"][0]
    return (route["distance"], route["duration"])
```

```python
#finding the distance from origin to all locations for that day
dist_array = []
for i , r in fifth_feb.iterrows():
    point1 = {"lat": r["lat"], "lon": r["lon"]}
    for j, o in fifth_feb[fifth_feb.index != i].iterrows():
        point2 = {"lat": o["lat"], "lon": o["lon"]}
        dist, duration = get_distance(point1, point2)
        #dist = geodesic((i_lat, i_lon), (o["CapitalLatitude"], o["CapitalLongitude"])).km
        dist_array.append((i, j, duration, dist))
```

```python
#renaming columns
distances_df = pd.DataFrame(dist_array,columns=["origin","destination","duration(s)","distnace(m)"])
distances_df.head()
```

```python
Fifth_Feb = distances_df[:5]
```

```python
#merge datasets to get more meaninful data
distances_df = Fifth_Feb.merge(data[["city"]], left_on = "origin", right_index=True).rename(columns={"city":"origin_name"})
distances_df = Fifth_Feb.merge(data[["city"]], left_on = "destination", right_index=True).rename(columns={"city":"destination_name"})
distances_df
```

```python
Fifth_Feb = distances_df.copy()
```

```python
#convert from metres to kilometres
Fifth_Feb['distance(km)'] = Fifth_Feb['distnace(m)']/1000
```

```python
# turn the first three columns of the dataframe into the list of tuples
dist_list = list(Fifth_Feb[["origin","destination","duration(s)"]].sort_values(by=["origin","destination"]).to_records(index=False))
dist_list[:5] + ["..."] + dist_list[-5:]
```

```python
length = Fifth_Feb.shape[0]
```

```python
#initializing travelling salesman for shortest route
fitness_dists = mlrose.TravellingSales(distances = dist_array)
```

```python
#defining travelling salesperson
problem_fit = mlrose.TSPOpt(length = length, fitness_fn = fitness_dists,
                            maximize=False)
```

```python
#genetic algorithm for finding shortest route
mlrose.genetic_alg(problem_fit, random_state = 2)
```

```python
#storing results 
best_state, best_fitness = mlrose.genetic_alg(problem_fit, random_state = 2)
```

```python
#displaying results of optimal(shortest) route using time(secs)
print(f"The best state found is: {best_state}, taking {best_fitness} ({str(dt.timedelta(seconds=best_fitness))})")
```

```python
orders = {city: order for order, city in enumerate(best_state)}
orders
```

```python
Fifth_Feb["order"] = Fifth_Feb.index.map(orders)
Fifth_Feb = Fifth_Feb.sort_values(by="order")
Fifth_Feb
```

```python
Fifth_Feb["next_city"] = Fifth_Feb["destination_name"].shift(-1)

# the last connection is between the last city and the first one
Fifth_Feb.loc[Fifth_Feb["order"] == max(Fifth_Feb["order"]), "next_city"] = Fifth_Feb.loc[Fifth_Feb["order"] == min(Fifth_Feb["order"]), "destination_name"].values[0]
Fifth_Feb
```
