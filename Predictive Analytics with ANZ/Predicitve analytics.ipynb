```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
from datetime import datetime
print('Setup complete!')
```

```python
data = pd.read_csv('new_data.csv')
data.head()
```

```python
data=data.drop(columns='Unnamed: 0')
data.head()
```

```python
cols_to_drop = ['status','account','currency','country','customer_id','movement','merchant_id','transaction_id','first_name','merchant_suburb','extraction','date']
```

```python
data = data.drop(columns=cols_to_drop)
```

```python
data.head()
```

```python
data.shape
```

```python
X = data.drop(columns='amount')
y = data['amount']
```

```python
X = pd.get_dummies(X, columns=['txn_description','gender','merchant_state','day_of_week'])
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import r2_score, mean_squared_error , mean_absolute_error
import xgboost
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print('shape of X_train:',X_train.shape)
print('shape of X_test:',X_test.shape)
print('shape of y_train:',y_train.shape)
print('shape of y_test:',y_test.shape)
```

```python
Lr = Lasso(alpha=0.8)
Lr.fit(X_train,y_train)
Lr_pred_train = Lr.predict(X_train)
Lr_pred_test = Lr.predict(X_test)
```

```python
rf = RandomForestRegressor(max_depth=2,n_estimators=3)
rf.fit(X_train,y_train)
rf_pred_train = rf.predict(X_train)
rf_pred_test = rf.predict(X_test)
```

```python
xg = XGBRegressor(max_depth=2,n_estimators=10)
xg.fit(X_train,y_train)
xg_pred_train = xg.predict(X_train)
xg_pred_test = xg.predict(X_test)
```

```python
dt = DecisionTreeRegressor(max_depth=1)
dt.fit(X_train,y_train)
dt_pred_train = dt.predict(X_train)
dt_pred_test = dt.predict(X_test)
```

```python
print('the mean squared error of the lasso regression model is',np.sqrt(mean_squared_error(y_test,Lr_pred_test)))
print('the mean squared error of the xgboost regression model is',np.sqrt(mean_squared_error(y_test,xg_pred_test)))
print('the mean squared error of the random forest regression model is',np.sqrt(mean_squared_error(y_test,rf_pred_test)))
print('the mean squared error of the decision tree regression model is',np.sqrt(mean_squared_error(y_test,dt_pred_test)))
```

```python
h=pd.DataFrame({'Model':['lasso','xgboost','random forest','Decision tree'],
             'F1 Score':[np.sqrt(mean_squared_error(y_test,Lr_pred_test)),np.sqrt(mean_squared_error(y_test,xg_pred_test)),
                        np.sqrt(mean_squared_error(y_test,rf_pred_test)),np.sqrt(mean_squared_error(y_test,dt_pred_test))]})
h.sort_values(by='F1 Score', ascending=True)
```

```python
print('the mean absolute error of the lasso regression model is',mean_absolute_error(y_test,Lr_pred_test))
print('the mean absolute error of the xgboost regression model is',mean_absolute_error(y_test,xg_pred_test))
print('the mean absolute error of the random forest regression model is',mean_absolute_error(y_test,rf_pred_test))
print('the mean absolute error of the decision tree regression model is',mean_absolute_error(y_test,dt_pred_test))
```
